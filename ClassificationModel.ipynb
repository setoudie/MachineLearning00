# Classification - K Nearest Neighbors (KNN)

La classification est une technique d'apprentissage supervisé en machine learning (ML) qui consiste à **assigner une étiquette ou une catégorie à un ensemble de données d'entrée**. En d'autres termes, l'objectif de la classification est de prédire la classe à laquelle un nouvel exemple appartient, en se basant sur un ensemble d'exemples préalablement étiquetés (jeu de données d'entraînement).

Voici les étapes principales du processus de classification :

1. **Collecte des données** : On commence par collecter des données qui contiennent à la fois des caractéristiques (features) et des étiquettes (classes ou catégories cibles).
   
2. **Entraînement du modèle** : Un modèle de machine learning est entraîné sur ces données étiquetées. L'algorithme apprend les relations entre les caractéristiques et les étiquettes.
   
3. **Prédiction** : Une fois entraîné, le modèle peut être utilisé pour prédire l'étiquette des nouvelles données non étiquetées.

4. **Évaluation** : Le modèle est évalué à l'aide de métriques comme l'exactitude (accuracy), la précision, le rappel (recall), la F1-score, etc., en fonction de ses prédictions sur un jeu de données de test.

Les algorithmes de classification populaires incluent :
- **Régression logistique**
- **K-Nearest Neighbors (KNN)**
- **Arbres de décision**
- **Support Vector Machines (SVM)**
- **Naive Bayes**
- **Réseaux de neurones** 

La classification est utilisée dans de nombreux domaines tels que la détection de spams, la reconnaissance d'image, la détection de fraude, etc.

## 1. Importation des librairies

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
```

## 2.Generation des donnes

```python
X, y = make_blobs(n_samples=50, centers=2, n_features=2, random_state=19, cluster_std=1)

X_new = np.random.uniform(low=X.min(), high=X.max(), size=(10,2))

plt.scatter(X[:,0], X[:,1], c=y, cmap="bwr")
plt.scatter(X_new[:,0], X_new[:,1], c='g')
plt.show()
```

## 3. Definition et entrainement du model

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)
```

## 4. Prediction

```python
predictions = model.predict(X_new)

# Scatter plot des données originales
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.3, label='Données originales')

# Scatter plot des nouvelles données prédictes
plt.scatter(X_new[:, 0], X_new[:, 1], c=predictions, edgecolors='k', s=100, label='Prédictions')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
