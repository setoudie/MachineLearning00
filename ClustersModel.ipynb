# Clustering - KMeans 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
k = np.random.choice([2,3,4,5,6,7,8,9])
X, _ = make_blobs(n_samples=200, centers=k, n_features=2, cluster_std=0.5)
```

En trancant le graphique de la variance en fonction de la valeur de , on voit que les donnes peuvent etre repartis en 4 clusters.

```python
k_range = range(1,10)
k_inertias = []

for k in k_range:
    model = KMeans(n_clusters=k).fit(X)
    k_inertias.append(model.inertia_) # model.inertia_ --> variance du model pour un k precis

plt.plot(k_range, k_inertias, lw=5, c='g', label='elbow courbe')
plt.xlabel('Valeurs de k')
plt.ylabel('Variance')
plt.show()
```
